```python
graph = {
    'O': [('Z', 71), ('S', 151)],
    'Z': [('A', 75), ('O', 71)],
    'S': [('F', 99), ('R', 80), ('A', 140), ('O', 151)],
    'A': [('T', 118), ('S', 140), ('Z', 75)],
    'F': [('B', 211), ('S', 99)],
    'R': [('P', 97), ('C', 146), ('S', 80)],
    'T': [('L', 111), ('A', 118)],
    'B': [('U', 85), ('G', 90), ('F', 211), ('P', 101)],
    'P': [('B', 101), ('C', 138), ('R', 97)],
    'C': [('D', 120), ('R', 146), ('P', 138)],
    'L': [('T', 111), ('M', 70)],
    'U': [('B', 85), ('V', 142), ('H', 98)],
    'G': [('B', 90)],
    'D': [('C', 120), ('M', 75)],
    'M': [('L', 70), ('D', 75)],
    'V': [('I', 92), ('U', 142)],
    'H': [('U', 98), ('E', 86)],
    'I': [('N', 87), ('V', 92)],
    'E': [('H', 86)],
    'N': [('I', 87)]
}

visited = {}
queue = []

def BFS(graph, src, dest):
    queue.append((src, 0, None))

    while queue:
        curr_node, curr_cost, parent = queue.pop(0)

        if curr_node == dest:
            print_path(parent, curr_node)
            return curr_cost

        if curr_node not in visited:
            visited[curr_node] = parent  # Store the parent of the current node in the visited dictionary

        print("Current Node:", curr_node)
        print("Visited Dictionary:", visited)
        print("Queue:", queue)
        print("--------")

        for child_node, child_cost in graph[curr_node]:
            if child_node not in visited:
                queue.append((child_node, curr_cost + child_cost, curr_node))

    return -1

def print_path(parent, node):
    path = [node]
    while parent:
        path.append(parent)
        parent = visited[parent]
    print(" -> ".join(reversed(path)))

source = input("Enter the Starting Node : ").capitalize()

destination = input("Enter the Ending Node : ").capitalize()

    
cost = BFS(graph,source,destination)
if cost != -1:
    print("Minimum Cost:", cost)
else:
    print("No path found.")
```

```python
graph = {
    'O': [('Z', 71), ('S', 151)],
    'Z': [('A', 75), ('O', 71)],
    'S': [('F', 99), ('R', 80), ('A', 140), ('O', 151)],
    'A': [('T', 118), ('S', 140), ('Z', 75)],
    'F': [('B', 211), ('S', 99)],
    'R': [('P', 97), ('C', 146), ('S', 80)],
    'T': [('L', 111), ('A', 118)],
    'B': [('U', 85), ('G', 90), ('F', 211), ('P', 101)],
    'P': [('B', 101), ('C', 138), ('R', 97)],
    'C': [('D', 120), ('R', 146), ('P', 138)],
    'L': [('T', 111), ('M', 70)],
    'U': [('B', 85), ('V', 142), ('H', 98)],
    'G': [('B', 90)],
    'D': [('C', 120), ('M', 75)],
    'M': [('L', 70), ('D', 75)],
    'V': [('I', 92), ('U', 142)],
    'H': [('U', 98), ('E', 86)],
    'I': [('N', 87), ('V', 92)],
    'E': [('H', 86)],
    'N': [('I', 87)]
}


def cal_path_cost(path):
    total_cost = 0
    for (node,cost) in path:
        total_cost += cost
    return total_cost, path[-1][0]

#checking utility function cal_path_cost()
path_chk = [('U', 85), ('G', 90), ('F', 211), ('P', 101)]
print("Checking Function :",cal_path_cost(path_chk)) #it will return last node and the cost till that node from start in a tuple
print("\n\n\n")

def UCS(graph,src,goal):
    visited = []
    queue = [[(src,0)]]
    while queue:
        queue.sort(key=cal_path_cost) # this will sort in asscending order 
        path = queue.pop(0) # accessing by least cost
        node = path[-1][0]  # accessing the last tuple and its first value, a node
        
        if node in visited:
            continue
        visited.append(node)
        if node == goal:
            return path
        else:
            adjcnt_nodes = graph.get(node,[])
            for (child_node, cost) in adjcnt_nodes:
                new_path = path.copy()
                new_path.append((child_node,cost))
                queue.append(new_path)

                

source = input("Enter the Starting Node : ").capitalize()

destination = input("Enter the Ending Node : ").capitalize()
                
                
path_way = UCS(graph,source,destination)
print("PathWay is : ",path_way)
print("Cost of Path : ",cal_path_cost(path_way))
```

# Iterative Deepening Depth first search


```python
graph = {
    'O': [('Z', 71), ('S', 151)],
    'Z': [('A', 75), ('O', 71)],
    'S': [('F', 99), ('R', 80), ('A', 140), ('O', 151)],
    'A': [('T', 118), ('S', 140), ('Z', 75)],
    'F': [('B', 211), ('S', 99)],
    'R': [('P', 97), ('C', 146), ('S', 80)],
    'T': [('L', 111), ('A', 118)],
    'B': [('U', 85), ('G', 90), ('F', 211), ('P', 101)],
    'P': [('B', 101), ('C', 138), ('R', 97)],
    'C': [('D', 120), ('R', 146), ('P', 138)],
    'L': [('T', 111), ('M', 70)],
    'U': [('B', 85), ('V', 142), ('H', 98)],
    'G': [('B', 90)],
    'D': [('C', 120), ('M', 75)],
    'M': [('L', 70), ('D', 75)],
    'V': [('I', 92), ('U', 142)],
    'H': [('U', 98), ('E', 86)],
    'I': [('N', 87), ('V', 92)],
    'E': [('H', 86)],
    'N': [('I', 87)]
}


def depth_limited_DFS(current_node, destination, depth, visited, current_path, graph):
    if depth == 0:
        return None
    
    visited.add(current_node)
    current_path.append(current_node)

    if current_node == destination:
        return current_path
    
    for neighbor, _ in graph.get(current_node, []):
        if neighbor not in visited:
            result = depth_limited_DFS(neighbor, destination, depth - 1, visited, current_path, graph)
            if result is not None:
                return result
    
    current_path.pop()
    return None


def iterative_deepening_DFS(start_node, end_node, graph):
    max_depth = 0
    
    while True:
        visited_nodes = set()
        current_path = []
        result_path = depth_limited_DFS(start_node, end_node, max_depth, visited_nodes, current_path, graph)
        
        if result_path is not None:
            return result_path
        
        max_depth += 1

start_node = input("Enter the Starting Node: ").capitalize()
end_node = input("Enter the Ending Node: ").capitalize()

final_path = iterative_deepening_DFS(start_node, end_node, graph)

if final_path:
    print("Pathway is:", final_path)
    print("Length of Path:", len(final_path) - 1)
else:
    print("No path found.")
```
